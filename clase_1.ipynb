```octave
pkg load image
ima = imread('rosa.jpg');
#descomponer en canales
R = ima(:,:,1);
G = ima(:,:,2);
B = ima(:,:,3);

imshow(ima); #muestra la imagen
```

```octave
#crear matriz para U,V y salida
[fil col] = size(R)#crea una matriz del tamao de R RED
U =zeros(fil, col); #crea una matriz con ceros
V = zeros(fil, col); #crea una matriz con ceros
salida = zeros(fil, col); #crea una matriz con ceros

#recorrer matriz con for

for f=1:1:fil
    for c=1:1:col
        U(fil, col) = (-0.14713*R(fil, col)) + (-0.28886*G(fil, col)) + (0.436*B(fil, col)); #YUV
        V(fil, col) = (0.615*R(fil, col)) + (-0.51498*G(fil, col)) + (-0.10001*B(fil, col)); #YUV
        
        if (U(fil, col)< 40) && (V(fil, col) > 40)
        salida(fil, col)=1;
        endif
    endfor
endfor
imshow(salida)
```

```octave
#convertir a escala de grises

gris = rgb2gray(ima);
imshow(gris)
```

```octave
#blanco y negro

bn = im2bw(ima,0.5);
imshow(bn)
```
