```octave
pkg load image
ima = imread('rosa.jpg');
#descomponer en canales
R = ima(:,:,1);
G = ima(:,:,2);
B = ima(:,:,3);

imshow(ima); #muestra la imagen
```

```octave
#crear matriz para U,V y salida
[fil col] = size(R)#crea una matriz del tamao de R RED
U =zeros(fil, col); #crea una matriz con ceros
V = zeros(fil, col); #crea una matriz con ceros
salida = zeros(fil, col); #crea una matriz con ceros

#recorrer matriz con for

for f=1:1:fil
    for c=1:1:col
        U(f,c) = (-0.14713*R(f, c)) + (-0.28886*G(f, c)) + (0.436*B(f, c)); #YUV
        V(f, c) = (0.615*R(f, c)) + (-0.51498*G(f, c)) + (-0.10001*B(f, c)); #YUV
        
        if (U(f, c)< 40) && (V(f, c) > 40)
        salida(f, c)=1;
        endif
    endfor
endfor
imshow(salida)
```

```octave
#convertir a escala de grises

gris = rgb2gray(ima);
imshow(gris)
```
